```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

# Folder paths
train_dir = 'CT_images/Train'
test_dir = 'CT_images/Test'

# Function to load images and labels
def load_images(directory):
    images = []
    labels = []
    for label_name in ['Stone', 'Normal']:
        folder = os.path.join(directory, label_name)
        for filename in os.listdir(folder):
            if filename.endswith(".jpg") or filename.endswith(".png"):
                img_path = os.path.join(folder, filename)
                img = cv2.imread(img_path)
                img = cv2.resize(img, (128, 128))  # Resize to a smaller size
                images.append(img)
                labels.append(0 if label_name == 'Stone' else 1)
    return np.array(images), np.array(labels)

# Load training and testing images
train_images, train_labels = load_images(train_dir)
test_images, test_labels = load_images(test_dir)

# Split training data for validation (80% training, 20% validation)
X_train, X_val, y_train, y_val = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Flatten images for the classifier
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_val_flat = X_val.reshape(X_val.shape[0], -1)
test_images_flat = test_images.reshape(test_images.shape[0], -1)

# Train a Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train_flat, y_train)

# Make predictions on training, validation, and test data
train_preds = clf.predict(X_train_flat)
val_preds = clf.predict(X_val_flat)
test_preds = clf.predict(test_images_flat)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, train_preds)
val_accuracy = accuracy_score(y_val, val_preds)
test_accuracy = accuracy_score(test_labels, test_preds)

# Print training, validation, and test accuracies
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Calculate Precision, Recall, F1 Score, and Confusion Matrix for test data
test_precision = precision_score(test_labels, test_preds)
test_recall = recall_score(test_labels, test_preds)
test_f1 = f1_score(test_labels, test_preds)
cm = confusion_matrix(test_labels, test_preds)

# Extract True Positives, False Positives, True Negatives, and False Negatives
TN, FP, FN, TP = cm.ravel()

# Print evaluation metrics
print(f"Precision: {test_precision:.2f}")
print(f"Recall: {test_recall:.2f}")
print(f"F1 Score: {test_f1:.2f}")
print("Confusion Matrix:")
print(cm)
print(f"True Positives: {TP}")
print(f"False Positives: {FP}")
print(f"True Negatives: {TN}")
print(f"False Negatives: {FN}")
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns

# Folder paths (update these with your local paths)
train_dir = 'CT_images/Train'
test_dir = 'CT_images/Test'

# Function to load images and labels
def load_images(directory):
    images = []
    labels = []
    for label_name in ['Stone', 'Normal']:
        folder = os.path.join(directory, label_name)
        for filename in os.listdir(folder):
            if filename.endswith(".jpg") or filename.endswith(".png"):
                img_path = os.path.join(folder, filename)
                img = cv2.imread(img_path)
                img = cv2.resize(img, (128, 128))  # Resize to a smaller size
                images.append(img)
                labels.append(0 if label_name == 'Stone' else 1)
    return np.array(images), np.array(labels)

# Load training and testing images
train_images, train_labels = load_images(train_dir)
test_images, test_labels = load_images(test_dir)

# Flatten images for the classifier
train_images_flat = train_images.reshape(train_images.shape[0], -1)
test_images_flat = test_images.reshape(test_images.shape[0], -1)

# Train a Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(clf, train_images_flat, train_labels, cv=skf)

# Train the model
clf.fit(train_images_flat, train_labels)

# Make predictions and evaluate
train_preds = clf.predict(train_images_flat)
test_preds = clf.predict(test_images_flat)

# Calculate metrics for the training data
train_accuracy = accuracy_score(train_labels, train_preds)
train_precision = precision_score(train_labels, train_preds)
train_recall = recall_score(train_labels, train_preds)
train_f1 = f1_score(train_labels, train_preds)

# Calculate metrics for the test data
test_accuracy = accuracy_score(test_labels, test_preds)
test_precision = precision_score(test_labels, test_preds)
test_recall = recall_score(test_labels, test_preds)
test_f1 = f1_score(test_labels, test_preds)

# Confusion matrix for test data
cm = confusion_matrix(test_labels, test_preds)

# Extract True Positives, False Positives, True Negatives, and False Negatives
TN, FP, FN, TP = cm.ravel()

# Print results
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Accuracy: {np.mean(cv_scores) * 100:.2f}%")
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"Training Precision: {train_precision:.2f}")
print(f"Test Precision: {test_precision:.2f}")
print(f"Training Recall: {train_recall:.2f}")
print(f"Test Recall: {test_recall:.f}")
print(f"Training F1 Score: {train_f1:.2f}")
print(f"Test F1 Score: {test_f1:.2f}")
print("Confusion Matrix:")
print(cm)
print(f"True Positives: {TP}, False Positives: {FP}, True Negatives: {TN}, False Negatives: {FN}")

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True,
            xticklabels=['No Stone', 'Stone'], yticklabels=['No Stone', 'Stone'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Learning Curves
def plot_learning_curve(model, X, y, X_test, y_test):
    train_sizes = np.linspace(0.1, 0.9, 10)  # Ensure train sizes are between 0.1 and 0.9
    train_scores = []
    test_scores = []

    for train_size in train_sizes:
        X_train_subset, _, y_train_subset, _ = train_test_split(X, y, train_size=train_size, random_state=42)
        model.fit(X_train_subset, y_train_subset)
        train_scores.append(model.score(X_train_subset, y_train_subset))
        test_scores.append(model.score(X_test, y_test))

    plt.plot(train_sizes, train_scores, label='Training score')
    plt.plot(train_sizes, test_scores, label='Validation score')
    plt.title('Learning Curve')
    plt.xlabel('Training Size')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.show()

# Call the function to plot learning curves using the test data
plot_learning_curve(clf, train_images_flat, train_labels, test_images_flat, test_labels)
```

```python

# Print results
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Accuracy: {np.mean(cv_scores) * 100:.5f}%")
print(f"Training Accuracy: {train_accuracy * 100:.5f}%")
print(f"Test Accuracy: {test_accuracy * 100:.5f}%")
print(f"Training Precision: {train_precision:.5f}")
print(f"Test Precision: {test_precision:.5f}")
print(f"Training Recall: {train_recall:.5f}")
print(f"Test Recall: {test_recall:.5f}")
print(f"Training F1 Score: {train_f1:.5f}")
print(f"Test F1 Score: {test_f1:.5f}")
print("Confusion Matrix:")
print(cm)
print(f"True Positives: {TP}, False Positives: {FP}, True Negatives: {TN}, False Negatives: {FN}")
```
